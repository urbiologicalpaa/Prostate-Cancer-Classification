## Feature selection(select top K features)

```python
from sklearn.feature_selection import SelectKBest, chi2

# Split the data into features (X) and target (y)
x = t_data.drop(["class"], axis=1)  # Assuming "class" is the target column
y = t_data["class"]

# Apply feature selection using chi-squared test
k = 1000  # Number of top features to select
selector = SelectKBest(chi2, k=k)
x_selected = selector.fit_transform(x, y)

# Get the indices of the selected features
selected_feature_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_feature_names = x.columns[selected_feature_indices]

# Create a new DataFrame with the selected features
selected_data = pd.DataFrame(x_selected, columns=selected_feature_names)
selected_data["class"] = y.reset_index(drop=True) # Add the target column to the selected data
```

```python
selected_data
```

```python
column_names = selected_data.columns.tolist()

# Exclude the 'class' column
prediction_feature = [col for col in column_names if col != 'class']
targeted_feature='class'
y = selected_data[targeted_feature]  #my dependent variable

# Remove 'class' column from the DataFrame to create 'x'
x =selected_data[prediction_feature]    #independent variable
```

```python
y.value_counts()
```

Having imbalanced classes in the target column can potentially affect the performance of your model, especially if the minority class (in this case, class '0') is underrepresented. Imbalanced classes can lead to biased predictions, where the model may favor the majority class due to its higher prevalence in the training data.

One approach to address class imbalance is to use data resampling techniques such as SMOTE (Synthetic Minority Over-sampling Technique). SMOTE generates synthetic samples for the minority class to balance the class distribution. This can help improve the model's ability to learn patterns from the minority class and make more accurate predictions.

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
x_resampled
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
    x_resampled,y_resampled,
    test_size=0.3,
    random_state=42
)

# Scale the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate, cross_val_score
from sklearn.svm import SVC
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def model_building(model,x_train_scaled,x_test_scaled,y_train, y_test):
    """
    
    Model Fitting, Prediction And Other stuff
    return ('score', 'accuracy_score', 'predictions' )
    """
    
    model.fit(x_train_scaled,y_train)
    score = model.score(x_train_scaled,y_train )
    predictions = model.predict(x_test_scaled)
    accuracy = accuracy_score(predictions, y_test)
    
    return (score, accuracy, predictions)  
```

```python

def cm_metrix_graph(cm):

    sns.heatmap(cm,annot=True,fmt="d", cmap="Blues")
    
    plt.xlabel("Predicted")
    plt.ylabel("Actual")

    plt.show()
```

```python
models_list = {
    "LogisticRegression" :  LogisticRegression(solver='liblinear'),
    "RandomForestClassifier" :  RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=5),
    "DecisionTreeClassifier" :  DecisionTreeClassifier(criterion='entropy', random_state=0),
    "SVC" :  SVC(),
}
```

```python
df_prediction = []
confusion_matrixs = []
df_prediction_cols = [ 'model_name', 'score', 'accuracy_score' , "accuracy_percentage"]

for name, model in zip(list(models_list.keys()), list(models_list.values())):
    
    (score, accuracy, predictions) = model_building(model, x_train_scaled, x_test_scaled, y_train, y_test )
    
    print("\n\nClassification Report of '"+ str(name), "'\n")
    
    print(classification_report(y_test, predictions))

    df_prediction.append([name, score, accuracy, "{0:.2%}".format(accuracy)])
    # For Showing Metrics
    confusion_matrixs.append(confusion_matrix(y_test, predictions))
    
        
df_pred = pd.DataFrame(df_prediction, columns=df_prediction_cols)
```

```python
plt.figure(figsize=(6, 4))
# plt.title("Confusion Metric Graph")


for index, cm in enumerate(confusion_matrixs):
    cm_metrix_graph(cm) # Call the Confusion Metrics Graph
    plt.tight_layout(pad=True)
   
```

```python
from sklearn.model_selection import cross_val_score

def k_fold_cv(models_list, x_resampled,y_resampled, cv=5):
    """
    Perform k-fold cross-validation for multiple models.
    
    Args:
        models (dict): A dictionary containing the models to evaluate.
        X (array-like): The feature matrix.
        y (array-like): The target variable.
        cv (int): The number of folds for cross-validation.
        
    Returns:
        dict: A dictionary containing the evaluation results for each model.
    """
    results = {}
    print("K-Fold Cross-Validation Results:")
    print("===============================")
    
    for model_name, model in models_list.items():
        # Perform cross-validation
        cv_scores = cross_val_score(model,x_resampled,y_resampled, cv=cv)
        
        # Store the results
        results[model_name] = cv_scores
        
        mean_accuracy = cv_scores.mean()
        std_deviation = cv_scores.std()
        
        # Print model evaluation summary
        print(f"Model: {model_name}")
        print(f"   Mean Accuracy: {mean_accuracy:.4f}")
        print(f"   Standard Deviation: {std_deviation:.4f}")
        print("   Individual Fold Scores:")
        
        for fold_idx, score in enumerate(cv_scores):
            print(f"      Fold {fold_idx+1}: {score:.4f}")
        
        print()
    
    return results
```

```python
results = k_fold_cv(models_list, x_resampled, y_resampled, cv=5)
```

```python
from  sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
def perform_grid_search(models_list,x_resampled,y_resampled, param_grid, cv=5):
    """
    Perform grid search with cross-validation for multiple models.
    
    Args:
        models (dict): A dictionary containing the models to evaluate.
        X (array-like): The feature matrix.
        y (array-like): The target variable.
        param_grid (dict): A dictionary of parameter grids for each model.
        cv (int): The number of folds for cross-validation.
        
    Returns:
        dict: A dictionary containing the best scores, estimators, and parameters for each model.
    """
    results = {}
    
    for model_name, model in models_list.items():
        # Perform grid search with cross-validation
        grid_search = GridSearchCV(model, param_grid[model_name], cv=cv)
        grid_search.fit(x_resampled,y_resampled)
        
        # Store the best score, estimator, and parameters
        best_score = grid_search.best_score_
        best_estimator = grid_search.best_estimator_
        best_params = grid_search.best_params_
        
        # Store the results
        results[model_name] = {
            'best_score': best_score,
            'best_estimator': best_estimator,
            'best_params': best_params
        }
        
        # Print the results for the current model
        print(f"Model: {model_name}")
        print(f"   Best Score: {best_score:.4f}")
        print("   Best Estimator:")
        print(best_estimator)
        print("   Best Parameters:")
        for param, value in best_params.items():
            print(f"      {param}: {value}")
        print()
    
    return results
```

```python
param_grid = {
    "LogisticRegression": {'C': [0.1, 1, 10],'penalty': ['l1', 'l2'],'solver': ['liblinear']},
    "RandomForestClassifier": {'n_estimators': [100, 200, 300],'max_depth': [5, 10],'bootstrap': [True, False],'max_depth': [40, 50],'max_features': ['auto', 'sqrt'],'min_samples_leaf': [1, 2],'min_samples_split': [2, 5],'n_estimators': [200, 400]},
    "DecisionTreeClassifier": {'max_depth': [5,10],'max_features': ['auto', 'sqrt', 'log2'],'min_samples_split': [2,3,4,5,6,7,8,9,10],'min_samples_leaf':[2,3,4,5,6,7,8,9,10]},
    "SVC": {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
}

results = perform_grid_search(models_list,x_resampled,y_resampled, param_grid, cv=5)
```
